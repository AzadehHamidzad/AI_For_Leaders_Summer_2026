# 1. Lesson on how to search for Python commands

Let's consider a few possible ways to learn about Python programming.  Let's suppose you want to learn how to produce a short summary of the information in your DataFrame.

1. Your **instructor** could provide the information.

You could be provided with a lesson about functions like info() and describe().  If you have a pandas DataFrame called df, then you can summarize its contents using df.info() or df.describe().  df.info() provides a list of column names with their counts and data types.  df.describe() will provide information such as the mean, min, max, standard deviation, and quantiles.  Thus:

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe()
```

In this describe() result, we see that the two columns A and B each have four elements.  The means and other statistics are shown.

2. You could look up the information on **Google**.

If I Google the question "how do I briefly summarize the contents of a dataframe using Python," I receive the following link (among others), which discusses the describe() command mentioned above:

https://www.w3schools.com/python/pandas/ref_df_describe.asp

It also provide the complete usage information:

dataframe.describe(percentiles, include, exclude, datetime_is_numeric)

It explains that "percentiles" is set by default to [0.25, 0.5, 0.75] but we could change that.  Let's try it!  Since there are three intervals here rather than four, it might be more meaningful to ask about a 33rd and 67th percentile rather than 25, 50, and 75.  We can use 1/3 for 0.33 and 2/3 for 0.67 to get the exact percentile values.

```python
df = pd.DataFrame([[1, 4], [2, 5], [3, 6], [4, 7]], columns = ['A', 'B'])
df.describe(percentiles = [1/3, 2/3])
```

Apparently, the 50% value (the median) stays even though we did not specifically request it.

8. What are the mean, median, and mode customer age in years?  (Rounding down to the next lower age.)
Are there any outliers?  (Customers with very large or very small ages, compared with the other ages?)

```python
today = pd.Timestamp.today()
df_bank_unique_account['Age'] = ((today - df_bank_unique_account["BirthDate"]).dt.days // 365)
```

```python
mean_age = int(df_bank_unique_account["Age"].mean())
median_age = int(df_bank_unique_account['Age'].median())
mode_age = int(df_bank_unique_account['Age'].mode()[0])

print("Mean:", mean_age)
print("Median:", median_age)
print("Mode:", mode_age)
```

```python
Q1 = df_bank_unique_account["Age"].quantile(0.25)
Q3 = df_bank_unique_account['Age'].quantile(0.75)

IQR = Q3 - Q1
lower = Q1 - 1.5 *IQR
upper = Q3 + 1.5 *IQR
outliers = df_bank_unique_account[
    (df_bank_unique_account["Age"] < lower)|
    (df_bank_unique_account["Age"] > upper)]
outliers
```

No outlier were detected using the IQR method.

```python
df2['AgeAtOpen'] = (df2["AccountOpened"] - df2['BirthDate']).dt.days //365
```

```python
df2 = df2.drop(columns=["AccountOpen"])
df2.head()
```

```python
import matplotlib.pyplot as plt

ax = plt.gca() # get an "Axes" object to draw on; gca stands for "get current Axes"
ax.scatter(df2["BirthDate"], df2["AccountOpened"]) # create a scatter plot based on these two dates
ax.set_ylabel("Account Opened") # label the y axis
ax.set_xlabel("Birth Date") # label the x axis
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(
    df2["BirthDate"].dt.year,
    df2["AgeAtOpen"]
)

plt.xlabel("Birth Year")
plt.ylabel("Age When Account Was Opened")
plt.title("Customer Age When Opening an Account by Birth Year")

plt.show()
```

The scatter plot does not show a strong relationship between a customer’s birth year and the age at which they opened their account. Customers seem to open bank accounts at a wide range of ages regardless of when they were born. The points are spread throughout the graph, and there is no clear upward or downward trend.

Most customers appear to have opened their accounts between the ages of about 10 and 70. However, there are a few unusual cases where customers appear to have opened accounts at very young ages. For example, some customers seem to have opened accounts when they were only a few years old. While this may be possible in special situations, it is uncommon and may indicate data quality issues that should be reviewed.

Overall, the graph suggests that birth year alone is not a strong predictor of the age at which a customer opens a bank account.

Choose any graph in the Introduction of Storytelling With Data (p. 1-17).  Using matplotlib to reproduce it in a rough way.
```python
import matplotlib.pyplot as plt

categories = ["Bored", "Not great", "OK", "Kind of\ninterested", "Excited"]
pre = [11, 5, 40, 25, 19]
post = [12, 6, 14, 30, 38]

x = range(len(categories))
width = 0.35

plt.figure(figsize=(8, 5))

plt.bar([i - width/2 for i in x], pre, width, label="PRE")
plt.bar([i + width/2 for i in x], post, width, label="POST")

plt.title("Pilot program was a success")
plt.ylabel("Percent of students")
plt.xticks(x, categories)
plt.ylim(0, 45)
plt.legend()

plt.show()
```
