# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

To make the viewer look at the last three months first, I would use preattentive attributes.

One way is to use color. I could make the earlier months light gray and make the last three months a darker color. This would make the last three months stand out quickly.

I could also use enclosure. For example, I could put a light gray shaded area behind the last three months. This would group those months together and show that they are the focus.

Another way is to use larger points. I could make the markers for the last three months bigger than the other points.

I could also add a short note near the last three months, such as “Last three months” or “Recent change.”

Overall, I would use color, size, shading, and a short note to guide the viewer’s attention to the last three months.

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
attributes = ['Engine power is less than expected', 'Tires make excessive noise while driving', 'Engine makes abnormal/excessive noise', 'Seat material concerns',
              'Excessive wind noise', 'Hesitation or delay when shifting', 'Bluetooth system has poor sound quality', 'Steering system/wheel has too much play', 
              'Bluetooth system is difficult to use', 'Front seat audio/entertainment/navigation controls']

percentages = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]
colors = ["indianred", "indianred", "indianred", "indianred", "indianred", "indianred", "indianred", "lightblue", "lightblue", "lightblue"]
plt.figure(figsize=(6, 6))
plt.barh(attributes, percentages, color=colors, height=0.71)

plt.gca().invert_yaxis()
plt.title('Top 10 design concerns', fontsize=17, loc='left', x=-0.2, y=1.06, ha='right', va='top')

plt.xticks([])
plt.yticks(fontsize=10)
plt.xlim(0, 16)
plt.tick_params(axis='x', length=0)
plt.tick_params(axis='y', length=0)

plt.gca().spines['right'].set_visible(False)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['left'].set_visible(False)
plt.gca().spines['bottom'].set_visible(False)

plt.text(0, -0.75, "concerns per 1,000", fontsize=10)

for i, v in enumerate(percentages):
    plt.text(v- 0.2 , i, f"{v:.1f}", color="white", va="center", ha="right", fontsize=10, fontweight="bold")
```
